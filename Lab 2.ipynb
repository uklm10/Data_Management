# Using Python, demonstrate a complete graph, where the number of vertices will be user-defined.

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
num_vertices = int(input("Enter the number of vertices: "))
complete_graph = nx.complete_graph(num_vertices)
nx.draw(complete_graph, with_labels=True, node_color='skyblue', node_size=500, font_size=14)
plt.title("Complete Graph with {} Vertices".format(num_vertices))
plt.axis('off')
plt.show()
```

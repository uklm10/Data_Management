# Using Python, draw a X-Y graph.

```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [3, 5, 7, 2, 1]
```

```python
# Line graph
plt.figure()
plt.plot(x, y, color='blue', label='Line Graph')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Graph Example')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Bar graph
plt.figure()
plt.bar(x, y, color='yellow', label='Bar Graph')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Bar Graph Example')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Histogram
plt.figure()
plt.hist(y, bins=5, color='green', edgecolor='black', label='Histogram')
plt.xlabel('Y-axis')
plt.ylabel('Frequency')
plt.title('Histogram Example')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Scatter plot
plt.figure()
plt.scatter(x, y, color='red', label='Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot Example')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Pie chart
plt.figure()
plt.pie(y, labels=x, autopct='%1.1f%%', startangle=90, colors=['yellow', 'orange', 'red', 'green', 'blue'])
plt.title('Pie Chart Example')
plt.axis('equal')
plt.show()
```
